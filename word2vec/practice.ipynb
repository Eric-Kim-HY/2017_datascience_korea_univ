```python
stem = nltk.stem.porter.PorterStemmer()
```

```python
f = open('./sample.txt')
text = f.readline()
tokens = nltk.word_tokenize(text)
n_tokens = len(tokens)
```

```python
freqdist = nltk.FreqDist(tokens)
```
